```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Interest Rates (%)': [1, 2, 3, 5, 7, 10, 12],
    'Doubling Time (months)': [72, 36, 24, 14, 10, 7, 5],
    'Future Value (Ksh)': [
        10000 * (1 + 0.01) ** (72 / 12),
        10000 * (1 + 0.02) ** (36 / 12),
        10000 * (1 + 0.03) ** (24 / 12),
        10000 * (1 + 0.05) ** (14 / 12),
        10000 * (1 + 0.07) ** (10 / 12),
        10000 * (1 + 0.10) ** (7 / 12),
        10000 * (1 + 0.12) ** (5 / 12)
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax1 = plt.subplots()

# Bar plot
ax1.bar(df.index, df['Doubling Time (months)'], color='skyblue', alpha=0.5, label='Doubling Time')
ax1.bar(df.index, df['Future Value (Ksh)'], color='grey', alpha=0.5, label='Future Value')

# Line plot (Interest Rates)
ax2 = ax1.twinx()
ax2.plot(df.index, df['Interest Rates (%)'], color='orange', marker='o', label='Interest Rates')

# Add trendline
z = np.polyfit(df.index, df['Interest Rates (%)'], 1)
p = np.poly1d(z)
ax2.plot(df.index, p(df.index), "orange", linestyle='--', label='Trendline')

# Set labels and legends
ax1.set_xlabel('Index')
ax1.set_ylabel('Doubling Time (months) / Future Value (Ksh)')
ax2.set_ylabel('Interest Rates (%)')

# Place legends outside the plot area
ax1.legend(loc='upper left', bbox_to_anchor=(1.1, 1))
ax2.legend(loc='upper right', bbox_to_anchor=(1.1, 0.9))

plt.title('Value of Ksh 10,000 computed over doubling time')
plt.xticks(df.index, df['Interest Rates (%)'])

plt.show()
```
